```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO("best.pt")
```

```python
model.predict("bus.jpg")
```

```python
def display_text(text, frame, position_tuple):
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1
    font_color = (255, 255, 255)  # White color
    font_thickness = 2
    text_position = (position_tuple[0], position_tuple[1]-10) # Coordinates for text position
    cv2.putText(frame, text, text_position, font, font_scale, font_color, font_thickness)
```

```python
import time
```

```python
cap = cv2.VideoCapture(0)
fps = 30

if not cap.isOpened():
    print("error opening camera")
    exit()

while True:
    start_time = time.time()
    ret, frame = cap.read()
    if not ret:
        print("Could not read frame")
        break
    try:
        result = model.predict(frame)
        result_arr = result[0].boxes.data
        item_dict = result[0].names
        for result in result_arr:
            cv2.rectangle(frame, (int(result[0]), int(result[1])), (int(result[2]), int(result[3])), (120, int(result[5])/80 * 255, (1-int(result[5]))/80 * 255), 5)
            display_text(item_dict[int(result[5])], frame, (int(result[0]), int(result[1])))
    except Exception as e:
        print("Error in getting predictions:", e)
        break
    cv2.imshow('Live Camera Feed', frame)
    elapsed_time = time.time() - start_time
    delay = max(0, 1/(fps - elapsed_time))
    if cv2.waitKey(int(delay*1000)) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
